```python
import pandas as pd

# Retrieve HTML table data
url = 'https://www.basketball-reference.com/leagues/NBA_2021_per_game.html'
html = pd.read_html(url, header = 0)
df2019 = html[0]


```

```python
# Data cleaning
raw = df2019.drop(df2019[df2019.Age == 'Age'].index)
raw
```

```python
raw.isnull().sum()
```

```python
df=raw.fillna(0)
```
